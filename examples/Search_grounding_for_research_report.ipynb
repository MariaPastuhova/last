```
train_y = clear_train_df.SalePrice
train_X = clear_train_df.drop(['SalePrice'], axis=1)
test_y = sample_df.SalePrice
test_X = clear_test_df
```

```
def get_mse(max_leaf_nodes, train_X, test_X, train_y, test_y):
    model = RandomForestRegressor(max_leaf_nodes=max_leaf_nodes, random_state=0)
    model.fit(train_X, train_y)
    preds_test = model.predict(test_X)
    mse = mean_squared_error(test_y, preds_test)
    return(mse)
```

```
max_leaf_nodes = 0
best_mse = 10**10
```
